```python
df_merged_only_num = df_merged.drop(['ID', 'LUGAR_DEL_HECHO' , 'FECHA_x' , 
       'Calle', 'Altura', 'Cruce', 'DirecciÃ³n Normalizada', 'pos x',
       'pos y', 'VICTIMA_x', 'ROL' , 'FECHA_FALLECIMIENTO'  ] , axis=1)
```

```python
'TIPO_DE_CALLE' , 'ACUSADO' , 'SEXO'  
```

```python
df_merged_only_num.replace("SD", 0, inplace=True)
```

```python
# Obtén variables ficticias para las columnas categóricas
df_dummies = pd.get_dummies(df_merged_only_num[['TIPO_DE_CALLE', 'ACUSADO', 'SEXO']], drop_first=True , dtype=int)

# Concatena las variables ficticias al DataFrame original
df_masdummies = pd.concat([df_merged_only_num, df_dummies], axis=1)

# Elimina las columnas originales categóricas si lo deseas
df_masdummies.drop(['TIPO_DE_CALLE', 'ACUSADO', 'SEXO'], axis=1, inplace=True)

# Muestra el DataFrame resultante
print(df_masdummies.head())
```

Guardo el dataframe con dummies 

```python
df_masdummies.to_csv('df_dummies.csv' , index=False)
```

```python
correlacion_matriz = df_masdummies.corr()
```

```python

plt.figure(figsize=(20, 15))  
sns.heatmap(correlacion_matriz, annot=True, cmap='plasma')
```

### Graficas

obtenemos los lugares con mas accidentes 

```python
Top_tipo_de_calle =df_homicidios_clean ['TIPO_DE_CALLE'].value_counts()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(y=Top_tipo_de_calle.index, x=Top_tipo_de_calle.values)
plt.xlabel('Cantidad / Amount')
plt.ylabel('Titulo / Title')
plt.title('25 títulos más comunes / The 25 most common titles')
plt.xticks(rotation=90)
plt.show()
```

Observaciones :
La principal observacion es que los hechos ocurren en su mayoria en las avenidas duplicando el valor del segundo lugar donde mas suceden los hechos que es la Calle y otra cosa a destacar es que la gral paz tiene el mismo numero de hechos que la suma de autopista 

```python
Top_Comuna =df_homicidios_clean ['COMUNA'].value_counts()
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(y=Top_Comuna.values, x=Top_Comuna.index)
plt.xlabel('COMUNA')
plt.ylabel('CANTIDAD')
plt.title('')
plt.xticks(rotation=90)
plt.show()
```
